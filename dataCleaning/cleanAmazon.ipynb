# Làm Sạch Dữ Liệu

```python
import pandas as pd
```

Đọc dữ liệu

```python
df = pd.read_csv('dataAmazon.csv')
df.head()
```

Kiểm tra và đếm số lượng giá trị NaN trong từng cột của DataFrame df

```python
df.isnull().sum()
```

Loại bỏ các dòng trùng lặp

```python
df_cleaned = df.drop_duplicates()
```

```python
df_cleaned.shape
```

Làm Sạch Cột 'brand':
- Loại bỏ phần chữ Top Brand:

```python
df_cleaned['brand'] = df_cleaned['brand'].str.replace('Top Brand:', '', regex=True)
df_cleaned.head()
```

Làm Sạch Cột 'price':
- Loại bỏ các dòng có giá trị không phải là số (chỉ giữ những dòng chứa số và ký tự $)

```python
df_cleaned = df_cleaned[df_cleaned['price'].str.contains('^\$\d+(\.\d{1,2})?$', na=False)]
```

- Loại bỏ dấu $ và chuyển đổi thành kiểu float

```python
df_cleaned['price'] = df_cleaned['price'].replace('[\$,]', '', regex=True).astype(float)
df_cleaned
```

Làm Sạch Cột 'ratings':
- Loại bỏ phần chữ " ratings" và chuyển đổi về kiểu số

```python
df_cleaned['ratings'] = df_cleaned['ratings'].replace('[\, ratings]', '', regex=True).astype(float)
df_cleaned.head()
```

Làm Sạch Cột 'rate':
- Làm sạch cột 'rate': chỉ giữ lại giá trị số

```python
df_cleaned['rate'] = df_cleaned['rate'].str.extract('(\d+\.\d+)').astype(float)
df_cleaned.head()
```

Kiểm tra xem có giá trị thiếu nào sau khi làm sạch và xử lý chúng

```python

df_cleaned.isnull().sum()
```

```python
df_cleaned
```

Kiểm Tra số dòng và cột dữ liệu 

```python
df_cleaned.shape
```

Loại bỏ các dòng có NaN ở cột 'brand'

```python
df_cleaned = df_cleaned.dropna(subset=['brand'])
df_cleaned
```

Những sản phẩm không rõ chất liệu vãi thì điền giá trị mặc định là "Không xác định"

```python
df_cleaned['fabricType'].fillna('Không xác định', inplace=True)
```

Những sản phẩm không rõ sản phẩm được sử dụng cách thức để mặc thì điền giá trị mặc định là "Không xác định"

```python
df_cleaned['closureType'].fillna('Không xác định', inplace=True)
```

Những sản phẩm không rõ quốc gia xuất xứ thì điền giá trị mặc định là "Không xác định"

```python
df_cleaned['countryOfOrigin'].fillna('Không xác định', inplace=True)
```

Những sản phẩm không rõ mô tả thì điền giá trị mặc định là "Không có"

```python
df_cleaned['describe'].fillna('Không có', inplace=True)
```

Những sản phẩm không rõ mô tả sản phẩm thì điền giá trị mặc định là "Không có"

```python
df_cleaned['productDescription'].fillna('Không có', inplace=True)
```

Những sản phẩm không bán được thì điền giá trị mặc định là "Không có"

```python
df_cleaned['ratings'].fillna('Không có', inplace=True)
```

Những sản phẩm không được đánh giá thì điền giá trị mặc định là "Không có"

```python
df_cleaned['rate'].fillna('Không có', inplace=True)
```

Kiểm tra dữ liệu còn bị thiếu sau khi xữ lý

```python
df_cleaned.isnull().sum()
```

Sau khi xử lý thì index của dòng dữ liệu đã thay đổi lung tung, ta cần reset index lại cho 
theo khuôn mẫu 

```python
df_cleaned.reset_index(drop=True, inplace=True)
```

Lưu trữ dữ liệu đã xử lý thành công với tên file dataAmazonClean.csv

```python
df_cleaned.to_csv('dataAmazonClean.csv')
```
