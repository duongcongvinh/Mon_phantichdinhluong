```python
import pandas as pd
from pymongo import MongoClient
import numpy as np
```

```python
# connect MongoDB
client = MongoClient('mongodb://localhost:27017/')
```

```python
# get database 
db = client['dbAmazon']
# get collection 
collection = db['datacrawler_copy']
```

```python
# get data from MongoDB
data = list(collection.find())
data
```

```python
# Tạo DataFrame từ danh sách
df = pd.DataFrame(data)
df.head()
```

```python
# xem kdl của từng cột trước khi xử lý 
df.dtypes
```

```python
# Xóa hai cột '_id' và 'productURL'  
df.drop(['_id', 'productURL'], axis=1, inplace=True)
```

```python
df.shape
```

```python
# Thay thế các giá trị không hợp lệ bằng NaN
df_cleaned = df
df_cleaned.replace(['', None], np.nan, inplace=True)
df_cleaned
```

```python
# Kiểm tra và đếm số lượng giá trị NaN trong từng cột của DataFrame df
df_cleaned.isnull().sum()
```

```python
# loại bỏ các dòng trùng lặp
df_cleaned = df.drop_duplicates()
```

```python
# sau khi loại bỏ các dòng trùng lặp
df_cleaned.shape
```

```python
# 1, Làm Sạch Cột 'brand':
# - Loại bỏ phần chữ Top Brand:
df_cleaned['brand'] = df_cleaned['brand'].str.replace('Top Brand:', '', regex=True)
df_cleaned.head()
```

```python
# 2, Làm Sạch Cột 'price':
# - Loại bỏ các dòng có giá trị không phải là số (chỉ giữ những dòng chứa số và ký tự $)
df_cleaned = df_cleaned[df_cleaned['price'].str.contains('^\$\d+(\.\d{1,2})?$', na=False)]
```

```python
# - Loại bỏ dấu $ và chuyển đổi thành kiểu float
df_cleaned['price'] = df_cleaned['price'].replace('[\$,]', '', regex=True).astype(float)
df_cleaned.head()
```

```python
# 3, Làm Sạch Cột 'ratings':
# - Loại bỏ phần chữ " ratings" và chuyển đổi về kiểu số
df_cleaned['ratings'] = df_cleaned['ratings'].replace('[\, ratings]', '', regex=True)
df_cleaned.head()
```

```python
# 4, Làm Sạch Cột 'rate':
# - Làm sạch cột 'rate': chỉ giữ lại giá trị số
df_cleaned['rate'] = df_cleaned['rate'].str.extract('(\d+\.\d+)').astype(float)
df_cleaned.head()
```

```python
# Kiểm tra xem có giá trị thiếu nào sau khi làm sạch và xử lý chúng
df_cleaned.isnull().sum()
```

```python
df_cleaned
```

```python
# Kiểm Tra số dòng và cột dữ liệu 
df_cleaned.shape
```

```python
# 5, Loại bỏ các dòng có NaN ở cột 'brand'
df_cleaned = df_cleaned.dropna(subset=['brand'])
df_cleaned.head()
```

```python
# 6, Những sản phẩm không rõ chất liệu vãi thì điền giá trị mặc định là "Không xác định"
df_cleaned['fabricType'].fillna('Unknown', inplace=True)
```

```python
# 7, Những sản phẩm không rõ sản phẩm được sử dụng cách thức để mặc thì điền giá trị mặc định là "Không xác định"
df_cleaned['closureType'].fillna('Unknown', inplace=True)
```

```python
# 8, Những sản phẩm không rõ quốc gia xuất xứ thì điền giá trị mặc định là "Không xác định"
df_cleaned['countryOfOrigin'].fillna('Unknown', inplace=True)
```

```python
# 9, Những sản phẩm không rõ mô tả sản phẩm thì điền giá trị mặc định là "Không có"
df_cleaned['productDescription'].fillna('Unknown', inplace=True)
```

```python
# 10, Những sản phẩm không rõ mô tả thì điền giá trị mặc định là "Không có"
df_cleaned['about'].fillna('Unknown', inplace=True)
```

```python
# 11, Những sản phẩm không bán được thì điền giá trị mặc định là "Không có"
df_cleaned['ratings'].fillna('Unknown', inplace=True)
```

```python
# 12, Những sản phẩm không được đánh giá thì điền giá trị 3
df_cleaned['rate'].fillna('3', inplace=True)
```

```python
# 13, Xóa cột customerssay  
df_cleaned.drop(['customerssay'], axis=1, inplace=True)
```

```python
# Kiểm tra dữ liệu còn bị thiếu sau khi xữ lý
df_cleaned.isnull().sum()
```

```python
# Sau khi xử lý thì index của dòng dữ liệu đã thay đổi lung tung, ta cần reset index lại cho 
# theo khuôn mẫu 
df_cleaned.reset_index(drop=True, inplace=True)
```

```python
# Đổi tên toàn bộ các cột về chữ thường
df_cleaned.columns = df_cleaned.columns.str.lower()
```

```python
# thêm cột product_id
df_cleaned['product_id'] = range(1,len(df_cleaned)+1)
```

```python
# gom nhóm brand và tạo brand_id
grouped_brands = df_cleaned.groupby('brand').size().reset_index(name='count')
grouped_brands['brand_id'] = range(1,len(grouped_brands)+1)
grouped_brands
```

```python
# Merge the grouped DataFrame with the original DataFrame on 'brand'
df_cleaned = df_cleaned.merge(grouped_brands[['brand', 'brand_id']], on='brand', how='left')

# Display the updated DataFrame
df_cleaned
```

```python
df_cleaned.dtypes
```

```python
# Lưu trữ dữ liệu đã xử lý thành công với tên file dataAmazonClean.csv
df_cleaned.to_csv('dataAmazonClean1_db.csv')
```

đưa dữ liệu lên postgres

```python
from sqlalchemy import create_engine

postgres_user = 'postgres' # Tên người dùng
postgres_password = '123456' # Mật khẩu
postgres_host = 'localhost' # Thay đổi từ 'localhost' thành tên container docker # chưa hoàn thành
postgres_port = '5432' # Cổng kết nối
postgres_db = 'DataAmazon' # Tên cơ sở dữ liệu
```

```python
# Kết nối tới PostgreSQL
# Thay thế 'username', 'password', 'hostname', 'database_name' bằng thông tin thực tế
engine = create_engine(f'postgresql://{postgres_user}:{postgres_password}@{postgres_host}:{postgres_port}/{postgres_db}')
```

```python
# đưa dữ liệu lên postgreSQL
# chia dữ liệu thành 5 phần tương ứng với 5 bảng đã tạo trong postgreSQL
df_brand = grouped_brands[['brand_id','brand']].copy()

df_product = df_cleaned[['product_id','productname','brand_id','price']].copy()

df_desproduct = df_cleaned[['product_id','about','productdescription']].copy()

df_productdetail = df_cleaned[['product_id','fabrictype','closuretype','countryoforigin']].copy()

dc_rate = df_cleaned[['product_id','ratings','rate']].copy()
```

```python
# đưa dữ liệu vào bảng brand trước vì bảng có khóa chính
df_brand.to_sql('brand', engine, if_exists='append', index=False)
```

```python
# đưa dữ liệu vào bảng product trước
df_product.to_sql('product', engine, if_exists='append', index=False)
```

```python
# đưa dữ liệu vào các bảng còn lại
df_desproduct.to_sql('desproduct', engine, if_exists='append', index=False)
df_productdetail.to_sql('productdetail', engine, if_exists='append', index=False)
dc_rate.to_sql('rate', engine, if_exists='append', index=False)
```
